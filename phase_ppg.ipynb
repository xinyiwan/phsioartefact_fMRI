```python
import pandas as pd 
import numpy as np 
import math

pulse = pd.read_csv('mri_pulse.csv').to_numpy()
peak = pd.read_csv('ppg_peak.csv').to_numpy()
```

```python
## Define the function that find the left ppg pulse

def find_left(array, value):
    array = np.asarray(array)
    idx = (np.abs(array - value)).argmin()
    if array[idx] > value:
        return idx-1
    else:
        return idx
```

```python
phase = []

for p in pulse:
    idx = find_left(peak,p)
    t1 = peak[idx][0]
    t2 = peak[idx+1][0]
    res = 2*math.pi*(p[0]-t1)/(t2-t1)
    print(t1,t2,p[0],p[0]-t1,t2-t1,res)
    phase.append(res)
```

```python
phase.to_csv('ppg_phase.csv', index = None)
```
