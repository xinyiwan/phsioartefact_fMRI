```python
import pandas as pd 
import numpy as np 
import math

resp_phase = pd.read_csv('resp_phase.csv').to_numpy()
card_phase = pd.read_csv('ppg_phase.csv').to_numpy()
mri_pulse = pd.read_csv('mri_pulse.csv').to_numpy()
```

## Define the order of regressors and output it

```python
m = 2
regressors = []
for i in range(len(mri_pulse)):
    t_reg = []
    for j in range(m):
        unit = [math.cos((j+1) * resp_phase[i]), math.sin((j+1) * resp_phase[i]), math.cos((j+1) * card_phase[i]), math.sin((j+1) * card_phase[i])]
        t_reg = np.concatenate((t_reg,unit))
    regressors.append(t_reg)
regressors = pd.DataFrame(regressors)
regressors.to_csv('regressors.csv', index = None)


```
